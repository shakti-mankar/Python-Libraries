```python
import pandas as pd 

s = pd.Series([23,45,12,67,43,56],
                index = (['jan','feb','march','apr','may','june']))

s.plot()
```

```python
s.plot(kind = 'barh', color='gray')
```

```python
s.plot(kind = 'pie')
```

```python
s.plot(kind='hist',color='pink')
```

```python
s.plot(kind='box')
```

```python
s.plot(kind='area', color='green')
```

```python
import matplotlib.pyplot as plot

s.plot(
kind='bar', # Chart type 
title= 'My Chart',  # Chart title 
xlabel = 'X Label', # x-axis label
ylabel = 'Y Label' ,# y-azis label 
figsize= (10,5), # (width,height) in inches
color = 'teal', # Single color or list of colors
colormap= 'viridis', # matplotlib color map
grid = True ,# Show legend 
rot = 45, # rotate x-axis labels 
fontsize=12, # axis tick font size 

)

plot.tight_layout()
plot.savefig('chart.png',dpi=150) # save to file 
plot.show()
```
