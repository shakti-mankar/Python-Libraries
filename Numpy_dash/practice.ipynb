```python
# ADDITION
arr_2 = np.array([10,20,30,40,50])

print(np.add(arr_2, 10 ))
```

```python
# SUBSTRACTION

arr_3 = np.array([10,20,30,40])

print(np.subtract(arr_3,2))
```

```python
# MULTIPLY 

arr_3 = np.array([10,20,30,40])

print(np.multiply(arr_3,2))
```

```python
# DIVIDE 


arr_3 = np.array([10,20,30,40])

print(np.divide(arr_3,2))


```

```python
# np.power


arr_3 = np.array([10,20,30,40])

print(np.power(arr_3,2))


```

```python
# SQUARE

arr_3 = np.array([10,20,30,40])

print(np.square(arr_3))
```

```python
# absolute

arr_3 = np.array([10,20,30,40])

print(np.abs(arr_3))
```

```python
# mod

arr_3 = np.array([10,20,30,40])

print(np.mod(arr_3 ,-11 ))
```
