```python
# 2. create a 5x5 matrix where the border contains 1s and the inside contains 0s

arr = np.zeros((5,5))

arr[0:1 , : ] = 1

arr[4: , :]  = 1

arr[1:: , ::4] = 1

print(arr)
```
