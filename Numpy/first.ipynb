```python
import numpy as np

import time

n = 100

# LISTS

list1 = list(range(n))
list2 = list(range(n))

start = time.time()

result_list = [list1[i] + list2[i] for i in range(n)]

list_time = time.time() - start

# Numpy Arrays

arr_1 = np.arange(n)
arr_2 = np.arange(n)

start = time.time()

result_arr = arr_1 + arr_2

numpy_time = time.time() - start

print(f"List time :  {list_time:.6f}, sec")
print(f"Numpy Time : {numpy_time:.6f} sec")
print(f"Numpy is approximately {list_time/numpy_time:.2f} x faster")
```
