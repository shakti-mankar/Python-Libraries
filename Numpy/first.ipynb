```python
import numpy as np

import time

n = 100

# LISTS

list1 = list(range(n))
list2 = list(range(n))

start = time.time()

result_list = [list1[i] + list2[i] for i in range(n)]

list_time = time.time() - start

# Numpy Arrays

arr_1 = np.arange(n)
arr_2 = np.arange(n)

start = time.time()

result_arr = arr_1 + arr_2

numpy_time = time.time() - start

print(f"List time :  {list_time:.6f}, sec")
print(f"Numpy Time : {numpy_time:.6f} sec")
print(f"Numpy is approximately {list_time/numpy_time:.2f} x faster")
```

```python
import numpy as np
arr = np.array([
        [1,2,3],
        [4,5,6]])

print(arr.flatten(order='C'  # 
                  ))

print(np.ravel(arr,order='A')) # memory 

print(np.ravel(arr,order='K')) # row wise


print(np.ravel(arr,order='F')) # column wise


# Flatten                 vs          Ravel
# 1. return copy                  1. returns views
# 2. Consume more memory          2. consume less memory
# 3. Slower                       3. Faster
# 4. Original unchange            4. Original will change 



```
